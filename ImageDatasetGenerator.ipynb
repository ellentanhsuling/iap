# Data Augmentation

```python
import numpy as np
import os
from keras.preprocessing.image import ImageDataGenerator
from PIL import Image  # Import the Image class from PIL

# Define the input and output folders
input_folder = 'drive/MyDrive/images/motorcycle'
output_folder = 'drive/MyDrive/images/AUGmotorcycle'
os.makedirs(output_folder, exist_ok=True)

# Create an instance of the ImageDataGenerator with desired augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,      # Randomly rotate images within the range of 20 degrees
    width_shift_range=0.1,  # Randomly shift images horizontally within 10% of the total width
    height_shift_range=0.1, # Randomly shift images vertically within 10% of the total height
    shear_range=0.2,        # Randomly apply shear transformation
    zoom_range=0.2,         # Randomly zoom in or out on images
    horizontal_flip=True,   # Randomly flip images horizontally
    fill_mode='nearest'     # Fill any newly created pixels after rotation or shifting
)

# Generate augmented images and save them to the output folder
for file_name in os.listdir(input_folder):
    # Construct the full path to the input image file
    input_image_path = os.path.join(input_folder, file_name)

    # Load the input image
    input_image = Image.open(input_image_path)
    input_image = input_image.convert('RGB')  # Convert to RGB if the image is in grayscale

    # Reshape the image to a 4D tensor with shape (1, height, width, channels)
    input_image = np.array(input_image)
    input_image = np.expand_dims(input_image, axis=0)

    # Generate augmented images
    augmented_images = datagen.flow(input_image, batch_size=1, save_to_dir=output_folder, save_prefix='aug_', save_format='jpg')

    # Generate a desired number of augmented images per input image
    num_augmented_images = 5
    for _ in range(num_augmented_images):
        augmented_image = augmented_images.next()

```

# STANDARDIZATION

```python
#For motorcycle

import os
from PIL import Image
import numpy as np

# Define the desired image size
image_size = (224, 224)

# Create a new folder to store the standardized images
output_folder = 'drive/MyDrive/images/NORMmotorcycle'
os.makedirs(output_folder, exist_ok=True)

# Load and resize the images
def resize_image(image_path):
    image = Image.open(image_path)
    image = image.resize(image_size)
    return image

# Normalize the pixel values
def normalize_image(image):
    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Normalize the pixel values using mean normalization
    mean = np.mean(image_array)
    std = np.std(image_array)
    normalized_image = (image_array - mean) / std

    return normalized_image

# Get the list of image files in the folder
input_folder = 'drive/MyDrive/images/AUGmotorcycle'
image_files = os.listdir(input_folder)

# Resize and normalize each image in the folder
for file_name in image_files:
    # Construct the full path to the image file
    image_path = os.path.join(input_folder, file_name)

    # Resize the image
    resized_image = resize_image(image_path)

    # Normalize the pixel values
    normalized_image = normalize_image(resized_image)

    # Save the standardized image to the output folder
    output_path = os.path.join(output_folder, file_name)
    Image.fromarray((normalized_image * 255).astype(np.uint8)).save(output_path)
```

# STANDARDIZATION WITH CONVERSION TO CSV

```python
import os
import csv
from PIL import Image
import numpy as np

# Define the desired image size
image_size = (200, 200)

# Create a new folder to store the standardized images
output_folder = 'drive/MyDrive/images/vehiclecsv'
os.makedirs(output_folder, exist_ok=True)

# Load and resize the images
def resize_image(image_path):
    image = Image.open(image_path)
    image = image.resize(image_size)
    return image

# Normalize the pixel values
def normalize_image(image):
    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Normalize the pixel values using mean normalization
    mean = np.mean(image_array)
    std = np.std(image_array)
    normalized_image = (image_array - mean) / std

    return normalized_image

# Get the list of image files in the first folder (motorcycle)
motorcycle_folder = 'drive/MyDrive/images/AUGmotorcycle'
motorcycle_files = os.listdir(motorcycle_folder)

# Get the list of image files in the second folder (car)
car_folder = 'drive/MyDrive/images/AUGcar'
car_files = os.listdir(car_folder)

# Initialize a list to store the image data and labels
image_data = []
labels = []

# Resize and normalize each image in the motorcycle folder
for file_name in motorcycle_files:
    # Construct the full path to the image file
    image_path = os.path.join(motorcycle_folder, file_name)

    # Resize the image
    resized_image = resize_image(image_path)

    # Normalize the pixel values
    normalized_image = normalize_image(resized_image)

    # Save the standardized image to the output folder
    output_path = os.path.join(output_folder, file_name)
    Image.fromarray((normalized_image * 255).astype(np.uint8)).save(output_path)

    # Flatten the image array to a 1D array
    flattened_image = normalized_image.flatten()

    # Add the flattened image array and label to the lists
    image_data.append(flattened_image)
    labels.append(1)  # Change label for motorcycle to 1

# Resize and normalize each image in the car folder
for file_name in car_files:
    # Construct the full path to the image file
    image_path = os.path.join(car_folder, file_name)

    # Resize the image
    resized_image = resize_image(image_path)

    # Normalize the pixel values
    normalized_image = normalize_image(resized_image)

    # Save the standardized image to the output folder
    output_path = os.path.join(output_folder, file_name)
    Image.fromarray((normalized_image * 255).astype(np.uint8)).save(output_path)

    # Flatten the image array to a 1D array
    flattened_image = normalized_image.flatten()

    # Add the flattened image array and label to the lists
    image_data.append(flattened_image)
    labels.append(0)  # Change label for honeydew to 0

# Convert the image data and labels lists to NumPy arrays
image_data = np.array(image_data)
labels = np.array(labels)

# Combine the image data and labels arrays into a single array
combined_data = np.column_stack((image_data, labels))

# Define the column names for the CSV file
column_names = ['pixel_' + str(i) for i in range(image_data.shape[1])] + ['label']

# Define the CSV file path
csv_file_path = 'drive/MyDrive/images/vehiclecsv/image_data.csv'

# Save the combined data as a CSV file
with open(csv_file_path, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(column_names)  # Write the column names to the CSV file
    writer.writerows(combined_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure x_train_reshaped is within the correct range and of the correct data type
# Assuming x_train_reshaped has values in the range [0, 1] due to normalization
x_train_display = (x_train_reshaped * 255).astype(np.uint8)

# Plotting the images
f, ax = plt.subplots(2, 5, figsize=(15, 6))  # Adjust figure size for better visibility

k = 0
for i in range(2):
    for j in range(5):
        # Display the image
        ax[i, j].imshow(x_train_display[k])
        ax[i, j].axis('off')  # Hide the axis
        k += 1

plt.tight_layout()
plt.show()
```

```python
model=tensorflow.keras.Sequential()
model.add(Conv2D(10 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (200,200,3)))
model.add(tensorflow.keras.layers.Flatten())
#model.add(keras.layers.Dense(1000,activation='relu',kernel_initializer='he_normal'))
#model.add(keras.layers.Dense(1000, activation='relu',kernel_initializer="he_normal"))
model.add(tensorflow.keras.layers.Dense(10 , activation='relu',kernel_initializer="he_normal"))
model.add(tensorflow.keras.layers.Dense(26, activation='softmax',kernel_initializer="he_normal"))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy']
)
```

```python
x_train_reshaped = np.array(x_train).reshape(-1, 200, 200, 3)
x_test_reshaped = np.array(x_test).reshape(-1, 200, 200, 3)

# Convert labels to numpy array and reshape to match the number of samples
y_train_reshaped = np.array(y_train)
y_test_reshaped = np.array(y_test)

history = model.fit(
    x_train_reshaped,
    y_train_reshaped,
    epochs=50,
    batch_size=16,
    validation_data=(x_test_reshaped, y_test_reshaped),
    verbose=1
)
```

```python
#Create a graph of loss vs epoch for train and test set.
#Write Your Code Below:

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
#Create a graph of accuracy vs epoch for train and test set.
#Write Your Code Below:

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```
